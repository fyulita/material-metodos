## Punto fijo
### Definiciones

+ $p$ es __punto fijo__ de una función $g$ si $g(p) = p$
+ Buscamos: $f(p)=0 \Longleftrightarrow g(p)=p $
+ Entonces, por ejemplo, $f(x)=0 \Longleftrightarrow \underbrace{f(x) + x}_{g(x)} = x$
+ Una vez hallada $g(x)$ a partir de $f(x)$ definimos la iteración de punto fijo: $$x_{n} = g(x_{n-1}), \qquad n=1,2,\ \ldots$$ Buscamos que la sucesión de valores $x_{0},x_{1},x_{2},\ \ldots\ $ generados por la iteración converja al punto fijo de $g$ (que a su vez es raíz de $f$)


### Teorema del Punto fijo
Sea $g(x)$ una función __continua__ en $[a, b]$ tal que $g(x) \in [a, b]$ para todo $x \in [a, b]$, suponiendo además que existe $g'$ en $(a,b)$ y que __existe una constante__ $k$}, con $0 < k < 1$ que cumple que $|g'(x)| \leq k$, para todo $ x \in (a, b)$.

Entonces, para cualquier $x_{0}$ en $[a,b]$, la sucesión $\{x_n\}_{n=0,\ 1,\ \ldots}$ definida por      $$ x_{n} = g(x_{n-1}), \quad n\geq 1,$$ converge al único punto fijo $p$ en $[a,b]$

Sean $y_1,\ \dots,\ y_n \in\mathbb{R}^3$ las ubicaciones de las naves estelares en el espacio (puntos verdes en la figura), y sea $h:[0,1]\to\mathbb{R}^3$ la función de trayectoria del Halcón Milenario, expresada como una trayectoria paramétrica en función del tiempo. Asumimos que la trayectoria es una **recta**, es decir $h(t) = at+b$, con $a,b\in\mathbb{R}^3$ (ver modelo en la figura). En el instante $t\in[0,1]$, Han Solo estará en la posición $h(t)$, y su **nivel de peligro**
está dado por:
\begin{equation}
A(t) \ =\ \sum_{i=1}^n {\frac{1}{\| h(t) - y_i\|_2}}
\end{equation}
Es decir, cada nave estelar aporta al nivel de peligro una cantidad que es inversamente proporcional a la distancia del Halcón Milenario a la nave estelar.  Si en algún instante $t\in[0,1]$ el nivel de peligro $A(t)$ es mayor o igual que un cierto **valor crítico** $C$, Han Solo deberá ser alertado para que active sus escudos y llegar a salvo a la Estrella de la Muerte.

## Definiciones

+ $n$ es la cantidad de naves estelares
+ $y_1,\ \dots,\ y_n \in\mathbb{R}^3$ las ubicaciones de las naves estelares
+  $h:[0,1]\to\mathbb{R}^3$ la función de trayectoria
+ $h$ es una recta: $h(t) = at+b$, con $a,\ b\in\mathbb{R}^3$.
+ El Halcón Milenario en el instante $t$ se encuentra en la posición $h(t)$
+ Nivel de peligro en instante $t$: $\displaystyle A(t) \ =\ \sum_{i=1}^n {\frac{1}{\| h(t) - y_i\|_2}}$
+ Cada nave estelar aporta al nivel de peligro $A(t)$ una cantidad que es inversamente proporcional a la distancia del Halcón Milenario a la nave


Veamos la derivada de la función del __nivel de peligro__


$$ A(t) 
= \sum_{i=1}^n {\frac{1}{\| h(t) - y_i\|_2}} 
= \sum_{i=1}^n {\sqrt{\frac{1}{\| h(t) - y_i\|_2^2}}}
= \sum_{i=1}^n {\big(\| h(t) - y_i\|_2^2\big)}^{-1/2}
= \sum_{i=1}^n {g_i(t)}
$$

donde $g_i(t) = m_i(t)^{-1/2}$, con $m_i(t) = \| h(t) - y_i\|_2^2$.

Luego,

$$
A'(t) = \sum_{i=1}^n {g'_i(t)}
$$

con $g'_i(t) = -\frac{1}{2} m_i(t)^{-3/2} m_i(t) = -\frac{1}{2} \frac{m'(t)}{m_i(t)^{3/2}}$

Recordemos que $a, b, y_i$ son vectores y $t$ es un escalar. Así, 
$ m_i(t) = \| h(t) - y_i\|_2^2 
= \|at + b - y_i\|_2^2 
= (at+b-y_i)^T(at+b-y_i) $
$ = (a^T a)t^2 + a^T(b-y_i)t + (b-y_i)^T at + \|b-y_i\|_2^2 $

$ = (a^T a)t^2 + 2a^T(b- y_i)t + \|b-y_i\|_2^2 $

Y su derivada:
$ m'_i(t) = 2(a^Ta)t + 2a^T(b-y_i) = 2a^T(at+b-y_i) = 2a^T(h(t) - y_i)$

Agrupando Todo:
$$
A'(t) = \sum_{i=1}^n {g'_i(t)} 
= \sum_{i=1}^n {-\frac{1}{2}\frac{2a^T(h(t) - y_i)}{\big(\| h(t) - y_i\|_2^2\big)^{3/2}}}
= -\sum_{i=1}^n {\frac{a^T(h(t) - y_i)}{\| h(t) - y_i\|_2^3}}
$$

 $$C=1, y_{1}=(2,3,-0.5), y_{2}=(2,0.5,0.5), a=(3,1,1), b=(1,1,-0.5)$$
**Nota:** Tener en cuenta las condiciones de convergencia de Newton-Raphson
 

```python
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Parámetros de la recta
a = np.array([3,1,1])
b = np.array([1,1,-0.5])

# Coordenadas de las naves estelares en R^{3 x n}
y = np.array([[2,3,-0.5], [2,0.5,0.5]])

# Cota para el nivel de aturdimiento
C = 1

# Función a encontrar la raíz
def f(t):
  return A(t) - C

# Derivada de F
def df(t):
  return dA(t)

# Función de nivel de aturdimiento
def A(t):
  suma = 0
  for nave in y:
    suma = suma + np.sum(1 / np.linalg.norm(a * t + b - nave))  
  
  return suma

# Derivada de A
def dA(t):
  suma = 0
  for nave in y:
    suma = suma + ((a * t + b - nave) / np.linalg.norm(a * t + b - nave) ** 3)
  
  return - a.T @ suma

print(dA(0))
```

```python
# NEWTON-RAPHSON
def NewtonRaphson(x0, tol, iteraciones):
  x_ant = x0
  contador = 0
  seq_x = [x0]

  while (abs(f(x_ant)) > tol and contador < iteraciones):
    x = x_ant - f(x_ant) / df(x_ant)
    
    seq_x.append(x)
    contador = contador + 1
    x_ant = x
  
  return (seq_x, x)
```

```python
x0 = 0.4 # COMPLETAR CON VALOR INICIAL
tol = 0.00000001 # PROBAR OTROS VALORES
iteraciones = 40

seq_nr, value_nr = NewtonRaphson(x0, tol, iteraciones)
its_nr = np.arange(1, len(seq_nr)+1)
err_nr = np.vectorize(abs)(np.vectorize(f)(seq_nr))

fig1 = plt.figure()
mlt.style.use('seaborn-notebook')
ax1 = plt.axes()
ax1.set_xlabel('k')
ax1.set_ylabel('x_k')
ax1.set_title('Aproximación iterativa de Newton-Raphson')
ax1.plot(its_nr, seq_nr, c='C2', label='Valor iterativo')
if abs(f(value_nr)) < tol:
  ax1.plot(its_nr, np.full(its_nr.shape, value_nr), dashes=[6,2], c='C0', label='Raíz estimada')
  ax1.legend(fancybox=True)

fig2 = plt.figure()
ax2 = plt.axes()
ax2.set_xlabel('k')
ax2.set_ylabel('e_k')
ax2.set_title('Error iterativo de Newton-Raphson')
ax2.plot(its_nr, err_nr, c='C2', label='Valor iterativo')
```
